```python
import pandas as pd
import numpy as np
import tensorflow as tf
import seaborn as sb
from tensorflow.keras.layers import *
from tensorflow.keras.regularizers import *
```

```python
df_train = pd.read_csv('Corona_NLP_train.csv', encoding='latin1')
df_test = pd.read_csv('Corona_NLP_test.csv', encoding='latin1')
```

```python
import re
```

```python
def regex_filter_tweet(sentence):
    tkn_filter = re.sub(r'http\S+', '', sentence) # regex to filter links
    tkn_filter = re.sub(r'[^\w\d\s\'\#]+', '', tkn_filter) # special filter for remove ponctuation while preserving tags
    tkn_filter = re.sub(r'[\x00-\x1f\x7f-\x9f]', ' ', tkn_filter) # remove control chars (\n,\r,\t...)
    tkn_filter = re.sub(' +', ' ', tkn_filter) # remove extra spaces in lines
    tkn_filter = tkn_filter.lower() # in favor of simplification use uncased sentences
    tags = []
    splited_tkns = tkn_filter.split(' ')
    tags = [t for t in splited_tkns if len(t) > 0 and t[0] == '#']
    return tkn_filter, len(splited_tkns), tags
```

```python
tokens = []
len_str = []
tags = []
for sentence in df_train['OriginalTweet']:
    tkns, words_len, tags_sent = regex_filter_tweet(sentence)
    len_str.append(words_len)
    tokens.append(tkns)
    tags.append(tags_sent)
df_train['filtered_tweet'] = tokens
df_train['words_count'] = len_str
df_train['tags'] = tags
```

```python
tokens = []
len_str = []
tags = []
for sentence in df_test['OriginalTweet']:
    tkns, words_len, tags_sent = regex_filter_tweet(sentence)
    len_str.append(words_len)
    tokens.append(tkns)
    tags.append(tags_sent)
df_test['filtered_tweet'] = tokens
df_test['words_count'] = len_str
df_test['tags'] = tags
```

```python
def get_frequency_from_corpus(corpus):
    freqs = {}
    for w in corpus:
        w = w.lower()
        if w not in freqs:
            freqs[w] = 1
        else: freqs[w] = 1+freqs[w]
    return freqs
```

```python
name = list(df_train['Sentiment'].unique())
fig = plt.figure(figsize=(16,12))

label_freq_all = {}

for i in range(len(name)):
    label_freq_all[str(name[i])] = {}
    subset = df_train[df_train['Sentiment']==name[i]]
    text = subset.filtered_tweet.values
    label_corpus = []
    for t in text:
        label_corpus += t.split(' ')
    label_corpus = set(label_corpus)
    label_freq_corpus = get_frequency_from_corpus(label_corpus)
    label_freq_all[str(name[i])].update(label_freq_corpus)
    wc = WordCloud(stopwords=STOPWORDS,background_color='black',collocations=False,width=2500,height=1800).generate_from_frequencies(label_freq_corpus)
    fig.add_subplot(3, 2, 1+i)
    plt.title(str(name[i]))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
plt.show()
```

```python
def get_ds_baseline(df, w2v):
    X = []
    for _,row in df.iterrows():
        fts = np.zeros(300)
        for word in row['filtered_tweet'].split(' '):
            try:
                fts += w2v.word_vec(word)
            except: pass
        X.append(fts)
    return np.array(X)
```

```python
X_train_bs = get_ds_baseline(df_train, word2vec_model)
X_test_bs = get_ds_baseline(df_test, word2vec_model)
```

```python
def baseline_model(in_dim=1024,n_labels=5):
    inp = Input(in_dim,)
    x = Dense(2048, activation='relu')(inp)
    x = Dropout(0.4)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(n_labels, activation='softmax')(x)
    model = tf.keras.Model(inp, x)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4), loss='categorical_crossentropy', metrics=['acc'])
    return model
```

```python
import tensorflow as tf
from transformers import GPT2Config, GPT2Tokenizer, GPT2Model
```

```python
def gtp2_clf(gpt2_model, in_dim=128, ff_dim=768, drop=0.2, n_labels=5, l2_reg=1e-5):

    ids = Input(shape=(in_dim,), dtype='int32')
    att_masks = Input(shape=(in_dim,), dtype='int32') 

    heads_cat = gpt2_model(ids, attention_mask=att_masks)[0] # concat of last layer attention heads
    mean_tokens = tf.reduce_mean(heads_cat, axis=1) # average of token embeddings in all heads

    ''' 
    ELU is faster to converge and more accurate 
    Dropout and l2 regularizer on kernel can help prevent overfitting in few-shot learning
    '''

    emb = Dense(ff_dim, kernel_regularizer=l2(l2_reg), name='embedded')(mean_tokens)
    emb = Activation('elu')(emb)
    emb = Dropout(drop)(emb) # Droping units improve generalization
    
    preds = Dense(n_labels, activation='softmax')(emb)
    model = tf.keras.Model([ids, att_masks], preds)

    return model
```

```python
def get_max_preds(y_preds_max, nb_labels):
    y_preds_max = np.argmax(y_preds_max, axis=1)
    y_preds = []
    for i in range(len(y_preds_max)):
        a = [False]*nb_labels
        a[y_preds_max[i]] = True
        y_preds.append(np.array(a))
    return np.array(y_preds)
```

```python
labels_names = ['Extremely Positive', 'Positive', 'Neutral', 'Negative', 'Extremely Negative']
```

```python
def get_metrics_results(y_test,y_preds):
    auc = sklearn.metrics.roc_auc_score(y_test, y_preds, multi_class='ovr')
    cm = sklearn.metrics.confusion_matrix(np.argmax(y_test, axis=1),np.argmax(np.rint(y_preds), axis=1))
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    f1 = sklearn.metrics.f1_score(np.argmax(y_test, axis=1),np.argmax(np.rint(y_preds), axis=1), average='macro')
    acc = sklearn.metrics.accuracy_score(np.argmax(y_test, axis=1),np.argmax(np.rint(y_preds), axis=1))
    result = {
        'acc':acc,
        'f1':f1,
        'auc_roc':auc,
        'cm':cm
    }
    return result

results = get_metrics_results(y_test,y_preds)
results_bs = get_metrics_results(y_test,y_preds_bs)

df_cm = pd.DataFrame(results['cm'], index = [i for i in labels_names],
                columns = [i for i in labels_names])
plt.figure(figsize=(4,3))
sb.heatmap(df_cm, annot=True, cmap='viridis')
plt.title('Confusion Matrix GPT2')
plt.show()
```
